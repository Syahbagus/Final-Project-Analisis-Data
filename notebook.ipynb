### Cleaning Data

Mengganti nama kolom pada data

```python
day_df = day_df.rename(columns={"yr": "year", "dteday": "date", "mnth": "month", "weathersit": "weather", "cnt": "total"})
hour_df = hour_df.rename(columns={"yr": "year", "dteday": "date", "mnth": "month", "weathersit": "weather", "cnt": "total"})

day_df
hour_df
```

Mengubah tipe data pada date menjadi datetime

```python
day_df["date"] = pd.to_datetime(day_df["date"])
hour_df["date"] = pd.to_datetime(hour_df["date"])

# Cek tipe data
day_df["date"].info()
hour_df["date"].info()
```

```python
# Konversi nilai pada kolom year
day_df["year"] = day_df["year"].replace({0: "2011", 1: "2012"})
hour_df["year"] = hour_df["year"].replace({0: "2011", 1: "2012"})
```

```python
# Konversi nilai pada kolom month
day_df["month"] = day_df["month"].replace({
    1: "Jan", 2: "Feb", 3: "Mar", 4: "Apr", 5: "May", 6: "Jun", 
    7: "Jul", 8: "Aug", 9: "Sep", 10: "Oct", 11: "Nov", 12: "Dec"
})

hour_df["month"] = hour_df["month"].replace({
    1: "Jan", 2: "Feb", 3: "Mar", 4: "Apr", 5: "May", 6: "Jun", 
    7: "Jul", 8: "Aug", 9: "Sep", 10: "Oct", 11: "Nov", 12: "Dec"
})
```

```python
# Konversi nilai pada kolom weekday
day_df["weekday"] = day_df["weekday"].replace({0: "Sun", 1: "Mon", 2: "Tue", 3: "Wed", 4: "Thu", 5: "Fri", 6: "Sat"})
hour_df["weekday"] = hour_df["weekday"].replace({0: "Sun", 1: "Mon", 2: "Tue", 3: "Wed", 4: "Thu", 5: "Fri", 6: "Sat"})
```

```python
# Konversi nilai pada kolom weather
day_df["weather"] = day_df["weather"].replace({1: "Clear", 2: "Mist & Cloudy", 3: "Light rain", 4: "Heavy rain"})
hour_df["weather"] = hour_df["weather"].replace({1: "Clear", 2: "Mist & Cloudy", 3: "Light rain", 4: "Heavy rain"})
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Pada musim apa penyewaan sepeda paling tinggi?

```python
color = {
    "Fall": "#8B4513",
    "Spring": "#E3D8C8",
    "Summer": "#E3D8C8", 
    "Winter": "#E3D8C8"
}
day_df["category"] = day_df["season"]

# Membuat grafik batang untuk season
plt.figure(figsize=(10, 5))
sns.barplot(
    x = "season",
    y = "total",
    data = day_df.sort_values(by="season", ascending=False),
    palette = color,
    hue = "category",
    legend = False,
    errorbar = None
)
plt.title("Total Counts by Season", fontsize=15)
```

Bisa dilihat dari grafik bahwa penyewaan tertinggi terjadi pada musim gugur, Dengan 1061129 penyewaan.

### Pertanyaan 2: Pada hari apa penyewaan paling tinggi?

```python
color={
    "Fri": "#8B4513",
    "Sun": "#E3D8C8",
    "Mon": "#E3D8C8",
    "Tue": "#E3D8C8",
    "Wed": "#E3D8C8",
    "Thu": "#E3D8C8",
    "Sat": "#E3D8C8"
}
day_df["category"] = day_df["weekday"]

# Membuat grafik batang untuk weekday
plt.figure(figsize=(10, 5))
sns.barplot(
    x = "weekday",
    y = "total",
    data = day_df.sort_values(by="weekday", ascending=False),
    order = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'],
    palette = color,
    hue = "category",
    legend = False,
    errorbar = None
)
plt.title("Total Counts by Weekday", fontsize=15)
```

Bisa dilihat dari grafik bahwa penyewaan tertinggi terjadi pada hari Jumat, Dengan jumlah 487790 penyewaan.

### Pertanyaan 3: Pada cuaca apa saja penyewaan paling tinggi?

```python
color={
    "Clear": "#8B4513",
    "Mist & Cloudy": "#E3D8C8",
    "Light rain": "#E3D8C8",
}
day_df["category"] = day_df["weather"]

# Membuat grafik batang untuk weather
plt.figure(figsize=(10, 5))
sns.barplot(
    x = "weather",
    y = "total",
    data = day_df.sort_values(by="weather", ascending=False),
    palette = color,
    hue = "category",
    legend = False,
    errorbar = None
)
plt.title("Total Counts by Season", fontsize=15)
```

Bisa dilihat dari grafik bahwa penyewaan tertinggi terjadi pada cuaca cerah, Dengan jumlah 2257952 penyewaan.

## Analisis Lanjutan

### RFM Analysis
RFM Analysis adalah metode analisis yang digunakan untuk mengevaluasi nilai pelanggan atau perilaku pengguna berdasarkan tiga dimensi utama:

- Recency: Kapan terakhir penyewaan dilakukan.
- Frequency: Seberapa sering penyewaan dilakukan
- Monetary: Total penyewaan sebagai proksi untuk nilai uang.

```python
# Menentukan tanggal terakhir
last_date = hour_df["date"].max()

# Menghitung Recency, Frequency, dan Monetary
rfm = hour_df.groupby("registered").agg(
    Recency=('date', lambda x: (last_date - x.max()).days),
    Frequency=('date', 'count'),
    Monetary=('total', 'sum')  
).reset_index()

# Menampilkan hasil RFM
print("RFM Analysis:")
rfm.head()
```
